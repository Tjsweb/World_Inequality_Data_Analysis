## Countries with least Life Expectancy over the years

```python
import plotly.graph_objs as go
def get_top6_countries(df, year):
    df_year = df[df['Year'] == year]
    return df_year.nsmallest(6, 'Life expectancy ')

years = df['Year'].unique().tolist()


fig = go.Figure()

df_initial = get_top6_countries(df, years[0])
fig.add_trace(go.Bar(x=df_initial['Country'], y=df_initial['Life expectancy '], name=str(years[0])))

fig.update_layout(title='Countries with the least Life Expectancy', xaxis_title='Country', yaxis_title='Life Expectancy (years)',
                  xaxis={'type': 'category'}, yaxis_range=[0, 100], barmode='group')

slider = dict(
    active=0,
    pad={"t": 50},
    steps=[],
)


for year in years:
    df_top6 = get_top6_countries(df, year)
    countries = df_top6['Country'].tolist()
    slider['steps'].append(dict(label=str(year), method="update", args=[{"x": [countries], "y": [df_top6['Life expectancy '].tolist()], "visible": [True]},
                                                                          {"title": str(year)}]))
    

fig.update_layout(sliders=[slider])

# Display the figure
fig.show()
```

#Hypothesis Testing

**Hypothesis 1:**

**Null hypothesis**: There is no significant difference in life expectancy between developed and developing countries.

**Alternative hypothesis**: There is a significant difference in life expectancy between developed and developing countries.

```python
from scipy.stats import ttest_ind

developed = df[df['Status'] == 'Developed']['Life expectancy ']
developing = df[df['Status'] == 'Developing']['Life expectancy ']

t_stat, p_val = ttest_ind(developed, developing)

print('t-statistic:', t_stat)
print('p-value:', p_val)
```

```python
# Significance level
alpha = 0.05

if p_val < alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

**Interpretation**: We have conducted a two-sample t-test to compare the life expectancy of developed and developing countries. The t-statistic obtained from the test was 29.805, and the p-value was 9.034e-171. We set the significance level (alpha) at 0.05. Since the p-value (9.034e-171) is less than the significance level (0.05), we reject the null hypothesis.

#Forecasting Future Life Expectancy of Developed and Developing Countries

```python
import plotly.graph_objs as go
import plotly.offline as pyo
from statsmodels.tsa.arima.model import ARIMA

df1 = df.copy()
df1['Year'] = pd.to_datetime(df1['Year'], format='%Y')

df_dev = df1[df1['Status'] == 'Developed']
df_dev_avg = df_dev.groupby('Year')['Life expectancy '].mean().reset_index()

df_devp = df1[df1['Status'] == 'Developing']
df_devp_avg = df_devp.groupby('Year')['Life expectancy '].mean().reset_index()

model_devp = ARIMA(df_devp_avg['Life expectancy '], order=(1,1,1))
model_devp_fit = model_devp.fit()
forecast_devp = model_devp_fit.forecast(steps=10)


model_dev = ARIMA(df_dev_avg['Life expectancy '], order=(1,1,1))
model_dev_fit = model_dev.fit()
forecast_dev = model_dev_fit.forecast(steps=10)

fig = go.Figure()

fig.add_trace(go.Scatter(x=df_dev_avg['Year'], y=df_dev_avg['Life expectancy '],
                    mode='lines+markers', name='Developed Historical Data'))

fig.add_trace(go.Scatter(x=df_devp_avg['Year'], y=df_devp_avg['Life expectancy '],
                    mode='lines+markers', name='Developing Historical Data'))

fig.add_trace(go.Scatter(x=pd.date_range(start='2016', periods=10, freq='Y'), y=forecast_dev,
                    mode='lines+markers', name='Developed Forecast'))

fig.add_trace(go.Scatter(x=pd.date_range(start='2016', periods=10, freq='Y'), y=forecast_devp,
                    mode='lines+markers', name='Developing Forecast'))

fig.update_layout(title='Life Expectancy Forecast for Developed and Developing Countries',
                   xaxis_title='Year',
                   yaxis_title='Life Expectancy')

fig.update_traces(hoverinfo='text', hovertext=[f'Year: {year}<br>Life Expectancy: {life_expectancy:.2f}'
                                               for year, life_expectancy in zip(df_dev_avg['Year'], df_dev_avg['Life expectancy '])],
                  selector=dict(mode='lines+markers', name='Developed Historical Data'))

fig.update_traces(hoverinfo='text', hovertext=[f'Year: {year}<br>Life Expectancy: {life_expectancy:.2f}'
                                               for year, life_expectancy in zip(df_devp_avg['Year'], df_devp_avg['Life expectancy '])],
                  selector=dict(mode='lines+markers', name='Developing Historical Data'))

fig.update_traces(hoverinfo='text', hovertext=[f'Year: {year}<br>Life Expectancy Forecast: {life_expectancy:.2f}'
                                               for year, life_expectancy in zip(pd.date_range(start='2016', periods=10, freq='Y'), forecast_dev)],
                  selector=dict(mode='lines+markers', name='Developed Forecast'))

fig.update_traces(hoverinfo='text', hovertext=[f'Year: {year}<br>Life Expectancy Forecast: {life_expectancy:.2f}'
                                               for year, life_expectancy in zip(pd.date_range(start='2016', periods=10, freq='Y'), forecast_devp)],
                  selector=dict(mode='lines+markers', name='Developing Forecast'))

pyo.iplot(fig)
```

**Interpretation:** Based on the time series forecasting graph, we can see that the life expectancy of developing countries is predicted to continue increasing steadily over the next 10 years. This supports our claim that there is a significant difference in life expectancy between developed and developing countries. Additionally, the forecasted line for developed countries is relatively flat, indicating that the life expectancy in these countries is already at a high level and is not expected to increase significantly in the future that is they have reached a saturation level.

**Hypothesis 2:**\
**Null Hypothesis**: There is no significant relationship between life expectancy and GDP, schooling, and income composition.

**Alternative Hypothesis**: There is a significant relationship between life expectancy and GDP, schooling, and income composition.

In this hypothesis testing we will be using **OLS Regression**.

Ordinary Least Squares (OLS) regression is a widely used statistical method for estimating the relationship between a dependent variable and one or more independent variables. The main principle of OLS is to minimize the sum of the squared differences between the observed dependent variable values and the predicted values obtained from the regression model.

```python
import statsmodels.api as sm
X = df[["GDP", "Schooling", "Income composition of resources"]]
y = df["Life expectancy "]

X = sm.add_constant(X)


model = sm.OLS(y, X).fit()

print(model.summary())
```

**Interpretation:** In the given OLS regression results:

**Dependent variable:** Life expectancy is the dependent variable, as we want to analyze the effect of GDP, schooling, and income composition of resources (independent variables) on life expectancy.

**Model**: The OLS model provides the best approximation of the true population regression line by minimizing the squared errors between the observed and predicted values

Coefficients and p-values:

**GDP**: The coefficient for GDP is 8.281e-05, with a p-value less than 0.05, indicating a significant positive relationship between GDP and life expectancy. As GDP increases, life expectancy is expected to increase by 8.281e-05 units, holding other factors constant.\

**Schooling**: The coefficient for schooling is 1.2307, with a p-value less than 0.05, indicating a significant positive relationship between schooling and life expectancy. As schooling increases, life expectancy is expected to increase by 1.2307 units, holding other factors constant.\

**Income composition of resources:** The coefficient for income composition of resources is 14.1886, with a p-value less than 0.05, indicating a significant positive relationship between income composition of resources and life expectancy. As the income composition of resources increases, life expectancy is expected to increase by 14.1886 units, holding other factors constant.

In conclusion, the OLS regression results show that there is a significant relationship between life expectancy and GDP, schooling, and income composition of resources. The coefficients for these independent variables indicate the direction and magnitude of their impact on life expectancy.
